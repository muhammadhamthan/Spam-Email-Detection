```python
import numpy as np
import pandas as pd
```

```python
#importing logistic regression model
from sklearn.linear_model import LogisticRegression # it is used for binary operation
```

```python
# importing accuracy score function...if acuracy is abv 90 means good but below means bad
from sklearn.metrics import accuracy_score
```

```python
# importing train test split function
from sklearn.model_selection import train_test_split
```

```python
# to convert all the string format into numeric
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
#importing the dataset
raw_mail_data = pd.read_csv('mail_data.csv')
raw_mail_data
```

```python
#checking the null values
raw_mail_data.isnull().sum()
#another method
#replacing all the null values with null string
mail_data = raw_mail_data.where(pd.notnull(raw_mail_data),'')
mail_data.info()
```

```python
#printing  the first 5 rows
mail_data.head()
```

```python
#how many rows and columns
mail_data.shape
```

**LABEL(DATA) ENCODING**

```python
#Label spam email as 0 and ham email as 1
mail_data.loc[mail_data['Category'] == 'ham', 'Category'] = 1
mail_data.loc[mail_data['Category'] == 'spam', 'Category'] = 0
mail_data
```

```python
#find the total number  of ham and spam email
mail_data['Category'].value_counts()
```

```python
#seperating the data as INPUT FEATURE and labelC(IT IS A  SUPERVISED LEARN)
X = mail_data['Message'] # INPUT FEATURE
Y = mail_data['Category'] # LABEL
```

```python
print(X)
```

```python
print(Y)
```

```python
#SPLITING THE DATA INTO TRAINING DATA AND TESTING DATA
#TRAINING DATA - X_TRAIN,Y_TRAIN
#TESTING DATA - X_TEST,Y_TEST
X_train , X_test, Y_train , Y_test = train_test_split(X,Y,test_size = 0.2,random_state =2)
print(X.shape,X_train.shape,X_test.shape)
```

```python
# transforming the text data to feature vectors(STRING) that can be used as input to the logistic regression...CONVERTING INTO THE NUMERIC FORMAT
feature_extraction = TfidfVectorizer(min_df = 1 , stop_words = "english" , lowercase  =True)
#min_df  = minimun document frequency
# stop_words = english words
```

```python
X_train_num = feature_extraction.fit_transform(X_train)
X_test_num = feature_extraction.transform(X_test)
```

```python
#convert Y_train values as integers
Y_train = Y_train.astype(int)
Y_test = Y_test.astype(int)
```

# training the model

```python
model = LogisticRegression()
```

```python
#training the logistic regression model with training data
model.fit(X_train_num,Y_train)
```

```python
#evaluating the trained model
X_train_prediction = model.predict(X_train_num)
```

```python
#finding the accuracy score for training data
training_data_accuracy = accuracy_score(Y_train,X_train_prediction)
print("accuracy score of training data",training_data_accuracy)
```

```python
#evaluating the testing model
X_test_prediction = model.predict(X_test_num)
```

```python
#FINDING THE ACCURACY SCORE FOR TESTING DATA
testing_data_accuracy = accuracy_score(Y_test,X_test_prediction)
print("accuracy score of testing data",testing_data_accuracy)
```

```python
#building a predictive System
input_data = ['Lol your always so convincing.']
input_data_num = feature_extraction.transform(input_data)
```

```python
#making prediction
prediction = model.predict(input_data_num)
if prediction == 1:
  print("ham email")
else:
  print("spam email")
```
